```python
df[['2019', '2020', '2021']] = df[['2019', '2020', '2021']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df
```

```python
df.count()
```

```python
print(df['Average'].describe())
```

```python
df['Change 2019-2020'] = ((df['2020'] - df['2019']) / df['2019']) * 100
df['Change 2020-2021'] = ((df['2021'] - df['2020']) / df['2020']) * 100
df
```

```python
def determine_trend(row):
    if row['Change 2019-2020'] > 0 and row['Change 2020-2021'] > 0:
        return 'Increasing'
    elif row['Change 2019-2020'] < 0 and row['Change 2020-2021'] < 0:
        return 'Decreasing'
    else:
        return 'Stable'
        
df['Trend'] = df.apply(determine_trend, axis=1)
df
```

```python
top_5 = df.nlargest(5, 'Average')
bottom_5 = df.nsmallest(5, 'Average')
top_5, bottom_5
```

```python
correlation = df['Average'].corr(df['Percentage share of state/UT (2018)'])
correlation
```

```python
plt.figure(figsize=(15,10))
for state in df['State/UT']:
    state_data = df[df['State/UT'] == state]
    plt.plot(['2019', '2020', '2021'], state_data[['2019', '2020', '2021']].values[0], label=state)
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.title('Crime Rates from 2019 to 2021 by State/UT')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['2019', '2020', '2021']])
plt.title('Box Plot of Crime Rates from 2019 to 2021')
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.set_index('State/UT')[['2019', '2020', '2021']], annot=True)
plt.title('Heatmap of Crime Rates from 2019 to 2021')
plt.show()
```

```python

df[['2019', '2020', '2021']].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Year-wise Comparison of Average Crime Rates')
plt.xlabel('Year')
plt.grid()
plt.ylabel('Average Crime Rate')
plt.show()
```
